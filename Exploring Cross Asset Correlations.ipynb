```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn_qqplot import pplot
```

```python
data = pd.ExcelFile("Data.xlsx")
```

```python
#Equity data
equity_level = pd.read_excel(data,"Equity_Index").set_index(["Date"])
equity_returns = pd.read_excel(data,"Equity_Returns").set_index(["Date"])
```

```python
equity_returns
```

```python
#FX data
fx_level = pd.read_excel(data,"FX_Index").set_index(["Date"])
fx_returns = pd.read_excel(data,"FX_Returns").set_index(["Date"])
```

```python
fx_returns
```

```python
#Commodity data
commodity_level = pd.read_excel(data,"Commodity_Index")
commodity_level.rename(columns = {commodity_level.columns[0]:"Date"},inplace = True)
commodity_level = commodity_level.set_index(["Date"])

commodity_returns = pd.read_excel(data,"Commodity_Returns")
commodity_returns.rename(columns = {commodity_returns.columns[0]:"Date"},inplace = True)
commodity_returns = commodity_returns.set_index(["Date"])
```

```python
commodity_level
```

```python
#IR data
ir_data = pd.read_excel(data,"IR_Level").set_index(["DATE"])
ir_returns = pd.read_excel(data,"IR_Returns").set_index(["DATE"])
```

```python
ir_data
```

## Exploring Cross Sectional Correlation

### Within Equity Asset Class

```python
plt.figure(figsize=(12,5))
sns.heatmap(equity_returns.corr(), cmap="YlGnBu", annot=True)
#plt.savefig('Figures/Correlation Analysis/Correlation_WithinEquity')
plt.show()
```

### Within FX Asset Class

```python
plt.figure(figsize=(12,5))
sns.heatmap(fx_returns.corr(), cmap="YlGnBu", annot=True)
#plt.savefig('Figures/Correlation Analysis/Correlation_WithinFX')
plt.show()
```

### Within Commodity Asset Class

```python
plt.figure(figsize=(8,5))
sns.heatmap(commodity_returns.corr(), cmap="YlGnBu", annot=True)
#plt.savefig('Figures/Correlation Analysis/Correlation_WithinCommodity')
plt.show()
```

### Within IR Asset Class

```python
plt.figure(figsize=(12,5))
sns.heatmap(ir_returns.corr(), cmap="YlGnBu", annot=True)
#plt.savefig('Figures/Correlation Analysis/Correlation_WithinIR')
plt.show()
```

### Across Asset Class

```python
combined = pd.concat([equity_returns,fx_returns,commodity_returns,ir_returns],axis = 1)
plt.figure(figsize=(20,12))
sns.heatmap(combined.corr(), cmap="YlGnBu", annot=False)
#plt.savefig('Figures/Correlation Analysis/Correlation_AcrossAssets')
plt.show()
```
